```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/game_data.csv')
```

```python
df_away = pd.DataFrame()
df_away['team'] = df.away
df_away['team_stats'] = df.away_stats
df_away['team_score'] = df.away_score
df_away['opp'] = df.home
df_away['opp_stats'] = df.home_stats
df_away['opp_score'] = df.home_score
df_away['home'] = 0
df_away['week'] = df.week
df_away['year'] = df.year
```

```python
df_home = pd.DataFrame()
df_home['team'] = df.home
df_home['team_stats'] = df.home_stats
df_home['team_score'] = df.home_score
df_home['opp'] = df.away
df_home['opp_stats'] = df.away_stats
df_home['opp_score'] = df.away_score
df_home['home'] = 1
df_home['week'] = df.week
df_home['year'] = df.year
```

```python
raw_data = pd.concat([df_away, df_home])
```
